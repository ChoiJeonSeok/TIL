```python
# 연령층과 월별 사고건수를 그룹화.
# 변수 이름 뜻 : 행_열
age_month = data.groupby(['가해자연령층', '발생월'])['사고건수'].sum().unstack()

# 그래프.
age_month.plot(kind='bar', figsize=(10, 6))
plt.title('가해자 연령층별 월별 교통사고 발생 건수')
plt.xlabel('월')
plt.ylabel('교통사고 발생 건수')
plt.legend(title='월', loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
age_month.head(10)
# 데이터를 살펴본 결과 불명은 각 연령층에 비하면 작은 부분
```

```python
# 불명 포함과 미포함
total_per_month_with = data.groupby('발생월')['사고건수'].sum()
total_per_month_unwith = data[data['가해자연령층'] != '불명'].groupby('발생월')['사고건수'].sum()

print(total_per_month_with)
total_per_month_unwith
```

```python
from scipy.stats import ttest_ind

# '가해자연령층'이 '불명'인 데이터가 제거되었을 때 유의미한 변화가 있는지 확인하고자 t-test 진행
# 귀무가설(H0): 가해자 연령층이 불명인 데이터와 불명이 제거된 데이터에서 월별 사고건수 평균이 같다.
# 대립가설(H1): 가해자 연령층이 불명인 데이터와 불명이 제거된 데이터에서 월별 사고건수 평균이 다르다.
t_statistic, p_value = ttest_ind(total_per_month_with, total_per_month_unwith)
print('p_value :', p_value, '\nt_statistic :', t_statistic)
```

```python
# p_value 값이 0.05보다 크다. 따라서 귀무가설을 기각할 수 없다. 
# t_statistic 절대값이 2 이하이다. 따라서 두 그룹의 
# 가해자연령층이 불명인 데이터는 제거되어도 사고건수에 대한 월별 차이는 유의하지 않다고 할 수 있다.
x_data = total_per_month_unwith.index
y_data = total_per_month_unwith.values

plt.plot(x_data, y_data, 'ro--')
plt.title("월별 사고건수")
plt.xlabel("발생월")
plt.ylabel("사고건수")
plt.grid(True)
plt.show()
```

```python
# 전체 data에 대해서 불명을 제거한 데이터와 t-test 진행
# '가해자연령층' 값이 '불명'인 행 제거한 DataFrame
# 귀무가설(H0): 가해자 연령층이 불명인 데이터와 불명이 제거된 데이터에서 사고건수의 data 평균이 같다.
# 대립가설(H1): 가해자 연령층이 불명인 데이터와 불명이 제거된 데이터에서 사고건수의 data 평균이 다르다.
from my_stat import TTest

data_no_unknown = data[data['가해자연령층'] != '불명']

# 두 그룹에 대한 t-test|
# 데이터, 실험군, 대조군, 분산(같다:True, 다르다:False)
col_names = list(data.columns[2:])
ttest = TTest(var=False)

for column in col_names:
    print(column)
    ttest.ttest( data_no_unknown[column],  data[column])
data_no_unknown
```

```python
# t-test 결과 가해자연령층에서 불명을 제거해도 전체 데이터에 영향을 주지 않는 것을 알 수 있다.
# 따라서 불명을 제거한 데이터를 반환한다.
data = data_no_unknown
data.to_csv('도로교통공단_가해운전자 연령층별 월별 교통사고 통계_2021년.csv', index=False, encoding='cp949')
```
