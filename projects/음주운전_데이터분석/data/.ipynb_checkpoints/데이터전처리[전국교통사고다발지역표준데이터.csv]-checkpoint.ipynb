```python
# 데이터 전처리 및 분석 기본 코드
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
%matplotlib inline
from matplotlib import font_manager

f_path = "C:/windows/Fonts/malgun.ttf"
font_manager.FontProperties(fname=f_path).get_name()
rc('font', family='Malgun Gothic')
```

```python
df = pd.read_csv('./done/전국교통사고다발지역표준데이터.csv', encoding='cp949')
```

```python
df.head(1)
```

```python
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
seoul_df = df[df['사고지역위치명'].str.contains('서울특별시')]
seoul_df = seoul_df[seoul_df['사고다발지역시도시군구'].str.contains('서울')]
seoul_df.head()
```

```python
seoul_df.info()
```

```python
seoul_df.loc[df['사고다발지역시도시군구'].str.startswith('서울'), '사고다발지역시도시군구'] = '서울특별시' 
seoul_df.head()
```

```python
print(seoul_df['사고다발지역시도시군구'].unique())
print(seoul_df['사고유형구분'].unique())
print(seoul_df['사고연도'].unique())
print(seoul_df['사고지역관리번호'].unique())
```

```python
seoul_df.to_csv('./result/서울교통사고다발지역표준데이터.csv', encoding='cp949')
```
