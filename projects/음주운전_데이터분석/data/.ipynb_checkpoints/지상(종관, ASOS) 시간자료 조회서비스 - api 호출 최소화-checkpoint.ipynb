```python
from datetime import datetime, timedelta

startDay = "20220101"
endDay = "20230101"

date_format = "%Y%m%d"

start_date = datetime.strptime(startDay, date_format)
end_date = datetime.strptime(endDay, date_format)

delta = end_date - start_date

# 시간으로 변환하기
delta_hours = delta.total_seconds() / 3600
delta_hours_as_int = int(delta_hours)
print(delta_hours_as_int)
```

```python
import xml.etree.ElementTree as ET
from bs4 import BeautifulSoup
import requests

result_dicts = []

key = "SGD3J7nUrX%2Bod%2B6SN3Fu%2BZB4N9DS13PXL4NV7jwYaFGTFjJSmN4Uk2eIr3ErirViBOIzt2L%2FiYtsg%2BItaPW6YA%3D%3D"

# 데이터를 받을 URL (pageNo 변수를 {}로 대체)
url_template = "https://apis.data.go.kr/1360000/AsosHourlyInfoService/getWthrDataList?serviceKey=" + key \
+ "&pageNo={}&numOfRows=1&dataType=XML&dataCd=ASOS&dateCd=HR" \
+ f"&startDt={startDay}&startHh=00&endDt={endDay}&endHh=23&stnIds=108"

# 페이지 번호를 바꾸어가며 데이터 가져오기
for i in range(1, delta_hours_as_int+1):
    url = url_template.format(i)
    
    # url로 request 보내기
    res = requests.get(url)
    if res.status_code == requests.codes.ok:  # HTTP 응답 코드가 정상인 경우 # 제대로 받았는지 check

        # HTML을 파싱한다. <item> 태그가 붙은 정보들을 findall 한다.
        soup = BeautifulSoup(res.text, 'html.parser')
        #ItemList = soup.findAll('item')
        #print(ItemList)

        # XML 파싱
        root = ET.fromstring(str(soup))

        # item 리스트 가져오기
        item_list = root.findall('body/items/item')

        # 추출할 태그 리스트 정의
        tag_list = ['tm', 'stnnm', 'rn','wd','hm','dsnw','ts']

        # 결과를 담을 딕셔너리 초기화
        result_dict = {tag: [] for tag in tag_list}

        # item 리스트에서 태그 추출하기
        for item in item_list:
            for tag in tag_list:
                # 값이 없는 경우 "Nan"으로 처리
                value = item.find(tag).text or "Nan"
                result_dict[tag].append(value)

        # 결과 출력
        result_dicts.append(result_dict)
        print(result_dict)
```

```python
import pandas as pd

# DataFrame 생성
df = pd.DataFrame(result_dicts)

# column명 설정
df.columns = ['tm', 'stnnm', 'rn','wd','hm','dsnw','ts']

# csv 파일로 저장
df.to_csv('weather2.csv', index=False, encoding='cp949')
```

```python
import csv

with open('weather2.csv', mode='w', encoding='cp949', newline='') as file:
    writer = csv.writer(file)
    # 첫 번째 행에 컬럼명 추가
    writer.writerow(['tm', 'stnnm', 'rn','wd','hm','dsnw','ts'])
    # 결과를 CSV 파일에 기록
    for result in result_dicts:
        # 값이 리스트인 경우 문자열로 변경
        for key in result:
            result[key] = result[key][0] if result[key] else ''
        # CSV 파일에 기록
        writer.writerow([result['tm'], result['stnnm'], result['rn'], result['wd'], result['hm'], result['dsnw'], result['ts']])
```
