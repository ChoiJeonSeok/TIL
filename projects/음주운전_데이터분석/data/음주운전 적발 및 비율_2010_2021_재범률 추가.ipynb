```python
# 데이터 전처리 및 분석 기본 코드
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
%matplotlib inline
from matplotlib import font_manager

f_path = "C:/windows/Fonts/malgun.ttf"
font_manager.FontProperties(fname=f_path).get_name()
rc('font', family='Malgun Gothic')
```

```python
df = pd.read_csv('./result/음주운전 적발 및 비율_2010_2021.csv', encoding='cp949')
```

```python
df.head()
```

```python
df.columns
```

```python
df = df.rename(columns={'년도': '연도'})
df
```

```python
df2 = df.iloc[:, list(range(9))+[11]]
df2
```

```python
total = df2.iloc[:, 3:8].sum(axis=1)
third_rate = total / df2['총합']

df2['재범률(3회 이상)'] = third_rate
df2
```

```python
df2 = df2.rename(columns={'재범률(총합 대비 2회 이상)':'재범률(2회 이상)'})
df2
```

```python
df2['재범률(2회 이상)'] = (df2['재범률(2회 이상)'] * 100).round(1).astype(str) + '%'
df2['재범률(3회 이상)'] = (df2['재범률(3회 이상)'] * 100).round(1).astype(str) + '%'
df2
```

```python
df3 = df2.copy()
df3 = df3.iloc[:, [0]+[9]+[10]]
df3
```

```python
import matplotlib.pyplot as plt
import pandas as pd


# 데이터 프레임을 표 이미지로 변환
fig, ax = plt.subplots()
ax.axis('off')
ax.table(cellText=df2.values, colLabels=df2.columns, cellLoc='center', loc='center')

# 이미지 저장
#plt.savefig('table_image.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 숫자 값으로 변환
df3['재범률(2회 이상)'] = df3['재범률(2회 이상)'].str.rstrip('%').astype(float)
df3['재범률(3회 이상)'] = df3['재범률(3회 이상)'].str.rstrip('%').astype(float)

# 그래프 그리기
plt.figure(figsize=(8, 6))  # 그래프 크기 설정

plt.plot(df3['연도'], df3['재범률(2회 이상)'], marker='o', label='재범률(2회 이상)')
plt.plot(df3['연도'], df3['재범률(3회 이상)'], marker='o', label='재범률(3회 이상)')

plt.xlabel('연도')
plt.ylabel('재범률 (%)')

plt.title('연도별 재범률')
plt.legend()

plt.grid(True)  
plt.show()
```
