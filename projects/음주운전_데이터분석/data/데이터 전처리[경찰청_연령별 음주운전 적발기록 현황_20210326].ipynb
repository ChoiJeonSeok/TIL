```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
%matplotlib inline
from matplotlib import font_manager

f_path = "C:/windows/Fonts/malgun.ttf"
font_manager.FontProperties(fname=f_path).get_name()
rc('font', family='Malgun Gothic')
```

```python
df = pd.read_csv('경찰청_연령별 음주운전 적발기록 현황_20210326.csv', encoding='cp949')
```

```python
df
```

```python
import numpy as np
df = df.rename(columns={'구분': '년도'})

# 데이터를 단순통합하는 것이므로 음주운전 재범자가 포함된 데이터의 연도를 축소하지 않음.
# df에 2021년도 데이터가 없으므로 NaN으로 채움
df.loc[11] = np.nan
df.loc[11]['년도'] = 2021
df
```

```python
data = pd.read_csv('./result/음주운전 적발 및 비율_2010_2021.csv', encoding='cp949')
```

```python
data
```

```python
# 년도를 기준으로 통합
merged_df = pd.merge(data, df, on='년도')
merged_df
```

```python
# 전체 운전자 대비 음주운전 적발 비율.
# 장롱면허도 있겠지만, 차량에 대한 접근성이 높은 현재로서는 본인이 희망하면 운전가능.
# 따라서 따로 처리하지 않고 적발자 비율 계산.

merged_df = merged_df.rename(columns={'적발자비율': '적발자비율(총합/운전면허 소지자 수)'})
merged_df
```

```python
merged_df.to_csv('음주운전 적발 및 비율_2010_2021.csv', index=False)
```
