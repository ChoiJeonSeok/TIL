```python
# '연도' 열의 데이터가 2010부터 2020까지인 행들만 선택하여 new_df에 저장
new_df = df.query('2010 <= 연도 <= 2020')

# 나머지 행들을 삭제
new_df = new_df.reset_index(drop=True)
new_df
```
