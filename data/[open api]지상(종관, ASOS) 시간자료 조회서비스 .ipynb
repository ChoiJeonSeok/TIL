```python
from datetime import datetime, timedelta

startDay = "20220101"
endDay = "20220102"

date_format = "%Y%m%d"

start_date = datetime.strptime(startDay, date_format)
end_date = datetime.strptime(endDay, date_format)

delta = end_date - start_date

# 시간으로 변환하기
delta_hours = delta.total_seconds() / 3600
delta_hours_as_int = int(delta_hours)
print(delta_hours_as_int)
```

```python
import xml.etree.ElementTree as ET
from bs4 import BeautifulSoup
import requests

result_dicts = []

key = "발급받은 키 입력"

 # 데이터를 받을 URL
url_template = "https://apis.data.go.kr/1360000/AsosHourlyInfoService/getWthrDataList?serviceKey=" + key \
+ "&pageNo={}&numOfRows=24&dataType=XML&dataCd=ASOS&dateCd=HR" \
+ f"&startDt={startDay}&startHh=00&endDt={endDay}&endHh=23&stnIds=108"

for i in range(1, delta_hours_as_int+1):
    url = url_template.format(i)

    # send request with url
    res = requests.get(url)
    if res.status_code == requests.codes.ok:  # check if the response was received correctly

        # parse the HTML, and find all the tags with the name "item"
        soup = BeautifulSoup(res.text, 'html.parser')
        root = ET.fromstring(str(soup))
        item_list = root.findall('body/items/item')
        
        # 추출할 태그 리스트 정의
        #tag_list = ['tm','ta', 'stnnm', 'rn','wd','wd','hm','dsnw','ts']

        # loop over the items in the list, and extract the desired tags
        # loop over the items in the list, and extract all tags
        for item in item_list:
            result_dict = {}
            for child in item:
                tag = child.tag
                value = child.text or "Nan"
                result_dict[tag] = value
            result_dicts.append(result_dict)

```

```python
import pandas as pd

# DataFrame 생성
df = pd.DataFrame(result_dicts)

# column명 설정
#df.columns = ['tm','ta', 'stnnm', 'rn','wd','hm','dsnw','ts']

# csv 파일로 저장
df.to_csv('weather_test1.csv', index=False, encoding='cp949')
```
